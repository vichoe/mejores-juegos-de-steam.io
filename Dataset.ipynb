```python
import pandas as pd

df = pd.read_csv("games.csv") 
df.head()
```

```python
columnas_utiles = [
    "AppID",
    "Name",
    "Release date",
    "Estimated owners",
    "Price",
    "Genres",
    "Positive",
    "Negative",
    "Average playtime forever"]

df = df[columnas_utiles]
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["Genres"]) # limpiar nulos
```

```python
df["Genres"] = df["Genres"].str.split(";")
df["Genres"] = df["Genres"].str[0]
```

```python
df = df[(df["Positive"] + df["Negative"]) > 0]

df["rating"] = df["Positive"] / (df["Positive"] + df["Negative"])
```

```python
df = df[df["Price"] >= 0] #poner precios validos
```

```python
df.head()
```

```python
#convertir rating en porcentaje
df["rating"] = df["rating"] * 100
df["rating"] = df["rating"].round(1)
df["rating"] = df["rating"].astype(str) + "%"
```

```python
df = df[df["Average playtime forever"] > 0]
```

```python
df.head(10)
```

```python
df = df.reset_index()
```

```python
df = df.drop(columns=['Estimated owners'])
```

```python
df = df.rename(columns={
    'index': 'AppID',
    'AppID': 'Name',
    'Name': 'Release date',
    'Release date': 'Estimated owners'
})
```
