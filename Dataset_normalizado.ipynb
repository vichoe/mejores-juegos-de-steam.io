```python
import pandas as pd
```

```python

df = pd.read_excel("games_limpio.xlsx")
df[['Min_Owners', 'Max_Owners']] = df['Estimated owners'].str.split(' - ', expand=True).astype(float)
df['Owners_Numeric'] = (df['Min_Owners'] + df['Max_Owners']) / 2
columnas_numericas = [
    'Price', 
    'Average playtime forever', 
    'Positive', 
    'Negative', 
    'Owners_Numeric'
]
df_final = df[['Name'] + columnas_numericas].copy()

for col in columnas_numericas:
    min_val = df_final[col].min()
    max_val = df_final[col].max()
    
    if max_val != min_val:
        df_final[f'{col}_Norm'] = ((df_final[col] - min_val) / (max_val - min_val)) * 100
    else:
        df_final[f'{col}_Norm'] = 0
df_final = df_final.dropna(subset=['Name'])
df_final = df_final.sort_values(by='Owners_Numeric', ascending=False)
display(df_final.head()) 

df_final.to_json("datos_radar.json", orient='records', indent=4)
df.head()
```
