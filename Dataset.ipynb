```python
columnas_utiles = [
    "AppID",
    "Name",
    "Release date",
    "Estimated owners",
    "Price",
    "Tags",
    "Negative", 
    "Score rank", 
    "Average playtime two weeks"]

df = df[columnas_utiles]
```

```python
columnas_nuevas = {
    'Negative': 'Positive', # negative representan los positive
    'Score rank': 'Negative' # Score rank representan los Negative
}

# Aplicar el cambio
df = df.rename(columns=columnas_nuevas)

print(df.columns)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["Tags", "Name"]) # limpiar nulos
```

```python
df["Tags"] = df["Tags"].str.split(";")
df["Tags"] = df["Tags"].str[0]
```

```python
df = df[(df["Positive"] + df["Negative"]) > 0]

df["rating"] = df["Positive"] / (df["Positive"] + df["Negative"])
```

```python
df = df[df["Price"] >= 0] #poner precios validos
```

```python
df.head()
```

```python
#convertir rating en porcentaje
df["rating"] = df["rating"] * 100
df["rating"] = df["rating"].round(1)
df["rating"] = df["rating"].astype(str) + "%"
```

```python
df = df[df["Average playtime two weeks"] > 0]
```
